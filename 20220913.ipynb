```python
x = range(0, 30)
y = [v*v for v in x]
y2 = [v*2*v for v in x]

plt.plot(x, y, 'mo', ms=15, mec='r')
plt.plot(x, y2, 'bx')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

x1 = np.array([0,1,2,3,])
y1 = np.array([13, 18 ,11 ,10])

x2 = np.array([0,1,2,3,])
y2 = np.array([32, 81 ,11 ,11])

x3 = np.array([0,1,2,3,])
y3 = np.array([320, 810 ,110 ,110])

x4 = np.array([0,1,2,3,])
y4 = np.array([30, 80 ,10 ,10])

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x3, y3)
ax4.plot(x4, y4)

plt.show()
```

```python
# plt.subplots(행, 열)
### plot, bar 그래프 그리기
x = range(0,15)
y = [v*v for v in x]
print(y)

fig, axs = plt.subplots(2,2, figsize=(10,10))
axs[0,0].plot(x, y, 'o')
axs[0,1].plot(x, y, '-')
axs[1,0].bar(x, y)
axs[1,1].hist(x, y)

plt.show()
```
