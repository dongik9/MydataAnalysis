```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 3)
plt.plot(x,y, "o")
plt.title("INCOME2")

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 4)
plt.plot(x,y, "o")
plt.title("INCOME2")

plt.suptitle("MY SHOP")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")

tips = sns.load_dataset("tips")  # 인터넷이 켜져 있어야 함.
tips

plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("day and size")
plt.show()
```

```python
### sns.boxplot() 
### sns.lmplot()
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y='tip', data=tips, size=7)
plt.show()
```
